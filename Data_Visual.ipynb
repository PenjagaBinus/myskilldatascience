```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("mpg")
```

```python
df.head()
```

```python
data_origin = df.origin.value_counts().to_dict()
```

```python
data_pie = data_origin.values()
data_labels = data_origin.keys()
colours = sns.color_palette('pastel')[0:5]
```

```python
plt.pie(x = data_pie,
        labels = data_labels,
        colors = colours,
        autopct = '%.0f%%')
plt.title("Origins")
plt.show()
```

```python
sns.histplot(data = df,
             x = 'mpg',
             bins = 30)
plt.show()
```

```python
sns.boxplot(data=df,
            x="mpg")
plt.show()
```

```python
sns.barplot(data = df,
              x = "origin",
              y = "mpg")
plt.show()
```

```python
sns.lineplot(data=df,
             x="model_year",
             y="mpg",
             hue="origin")
plt.show()
```

```python
sns.scatterplot(data = df,
                x = "displacement",
                y = "mpg")
plt.show()
```
